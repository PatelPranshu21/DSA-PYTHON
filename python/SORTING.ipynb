#SELECTION SORT
#TC->o(N^2)
#SC->o(1)

```python
nums=[5,8,6,3,2,4,1,9,3]
def sel_sort(nums):
    for i in range(0,len(nums)):
         min_index=i
         for j in range(i+1,len(nums)):
             if nums[j]<nums[min_index]:
                  min_index=j
         nums[i],nums[min_index]=nums[min_index],nums[i]
    return nums    
        
print(sel_sort(nums))
```

#BUBBLE SORT
#TC->o(N^2)
#SC->o(N)

```python
nums=[1223,3557,4235,315,313,64575,13428,4677]
n=len(nums)
def bubble_sort(nums):
    for i in range(n-2,-1,-1):
         is_swap=False
         for j in range(0,n-1):
              if nums[j]>nums[j+1]:
                  nums[j],nums[j+1]=nums[j+1],nums[j]
                  is_swap=True
    return nums
    if is_swap==False:
        return

print(bubble_sort(nums))        
```

#INSERTION SORT

#TC->o(N^2)
#SC->o(1)

```python
nums=[12,34,13,4,2,90,45,234]
def insert_sort(nums):
    n=len(nums)
    for i in range(1,n):
        key=nums[i]
        j=i-1
        while j>=0 and nums[j]>key:
            nums[j+1]=nums[j]
            j-=1
        nums[j+1]=key  
    return nums

print(insert_sort(nums))
```

#MERGE SORT

#TC->o(NlogN)
#SC->o(N)

```python
nums=[4,10,35,2,8,56,134,237,1,7,9]
def merge_array(left,right):
    result=[]
    n,m=len(left),len(right)
    i,j=0,0
    while i<n and j<m:
        if left[i]<=right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
    if i<n:
        while i<n:
            result.append(left[i])
            i+=1
    if j<m:
        while j<m:
            result.append(right[j])
            j+=1
    return result            
def merge_sort(arr):
    if len(arr)<=1:
        return arr
    mid=len(arr)//2
    left_array=arr[:mid]
    right_array=arr[mid:]
    left=merge_sort(left_array)
    right=merge_sort(right_array)
    return merge_array(left,right)

r=merge_sort(nums)
print(r)
```

#QUICK SORT


#TC->o(NlogN)

#SC->o(1)

```python
nums=[4,5,7,8,3,1,2,9]
def partition(nums,low,high):
    pivot=nums[low]
    i=low
    j=high
    while i<j:
        while nums[i]<=pivot and i<=high-1:
            i+=1
        while nums[j]>pivot and j>=low+1:
            j-=1
        if i<j:
            nums[i],nums[j]=nums[j],nums[i]
    nums[j],nums[low]=nums[low],nums[j] 
    return j  

def quick_sort(nums,low,high):
    if low<high:
        h=partition(nums,low,high)
        quick_sort(nums,low,h-1)
        quick_sort(nums,h+1,high)  

quick_sort(nums,0,len(nums)-1)
print(nums)
```
