# Binary Search

```python
def search(nums,target):
    n=len(nums)
    low=0
    high=n-1
    while low<=high:
        mid=(low+high)//2
        if nums[mid]==target:
            return mid
        elif nums[mid]>=target:
            high=mid-1
        else:
            low=mid+1
    return mid  
    
nums=[1,3,5,7]   
print(search(nums,4))                                                                         
```
